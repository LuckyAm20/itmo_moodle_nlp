```python
import networkx as nx
from networkx.algorithms.clique import find_cliques

G = nx.dense_gnm_random_graph(53, 656, seed=44)

density = round(nx.density(G), 3)
shortest_path_len = nx.shortest_path_length(G, source=3, target=47)
max_clique_size = max(len(clique) for clique in find_cliques(G))

print(density)
print(shortest_path_len)
print(max_clique_size)
```
