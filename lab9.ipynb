```python
import pandas as pd
import numpy as np

from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
train_df = pd.read_csv("train.csv")
val_df = pd.read_csv("val.csv")
```

```python
train_df["text"] = train_df["text"].fillna("").astype(str)
val_df["text"] = val_df["text"].fillna("").astype(str)

X_train = train_df["text"]
y_train = train_df["class"].astype(int)

X_val = val_df["text"]
y_val = val_df["class"].astype(int)
```

```python
model = Pipeline(
    [
        (
            "features",
            FeatureUnion(
                [
                    (
                        "word",
                        TfidfVectorizer(
                            analyzer="word",
                            ngram_range=(1, 2),
                            min_df=2,
                            max_df=0.98,
                            sublinear_tf=True,
                        ),
                    ),
                    (
                        "char",
                        TfidfVectorizer(
                            analyzer="char_wb",
                            ngram_range=(3, 5),
                            min_df=2,
                            sublinear_tf=True,
                        ),
                    ),
                ]
            ),
        ),
        ("clf", LinearSVC(C=1.0)),
    ]
)
```

```python
model.fit(X_train, y_train)
```

```python
val_pred = model.predict(X_val)
val_acc = accuracy_score(y_val, val_pred)
val_acc
```

```python
model2 = Pipeline(
    [
        (
            "features",
            FeatureUnion(
                [
                    (
                        "word",
                        TfidfVectorizer(
                            analyzer="word",
                            ngram_range=(1, 3),
                            min_df=2,
                            max_df=0.99,
                            sublinear_tf=True,
                        ),
                    ),
                    (
                        "char",
                        TfidfVectorizer(
                            analyzer="char_wb",
                            ngram_range=(2, 6),
                            min_df=2,
                            sublinear_tf=True,
                        ),
                    ),
                ]
            ),
        ),
        ("clf", LinearSVC(C=0.7)),
    ]
)

model2.fit(X_train, y_train)
val_pred2 = model2.predict(X_val)
val_acc2 = accuracy_score(y_val, val_pred2)
val_acc2
```

```python
best_model = model if val_acc >= val_acc2 else model2
best_val_acc = max(val_acc, val_acc2)
best_val_acc
```

```python
full_df = pd.concat([train_df, val_df], axis=0).reset_index(drop=True)
X_full = full_df["text"]
y_full = full_df["class"].astype(int)

best_model.fit(X_full, y_full)
```

```python
test_df = pd.read_csv("test.csv")
test_df["text"] = test_df["text"].fillna("").astype(str)
X_test = test_df["text"]
```

```python
test_pred = best_model.predict(X_test)
predictions = test_pred.astype(int).tolist()
predictions[:20], len(predictions)
predictions
```
