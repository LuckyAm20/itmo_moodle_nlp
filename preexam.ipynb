```python
import math

def ppmi(w0, w1, sentences):
  return max(math.log2(len(sentences) * sum([1 for w in sentences if w0 in w and w1 in w]) / sum([1 for w in sentences if w0 in w]) / sum([1 for w in sentences if w1 in w])), 0)
```

```python
print(ppmi("lines", "organization", sentences))
print(ppmi("article", "world", sentences))
print(ppmi("world", "organization", sentences))
```
