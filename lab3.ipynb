```python
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("Task_1_prepprocessed.csv")
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
y = df.iloc[:, 0]

x_body = df.iloc[:, 5]

x_train_a, x_test_a, y_train, y_test = train_test_split(
    x_body, y, test_size=0.25, random_state=94
)
```

```python
vectorizer_a = TfidfVectorizer()

X_train_a = vectorizer_a.fit_transform(x_train_a)
X_test_a = vectorizer_a.transform(x_test_a)

model_a = RandomForestClassifier(n_estimators=8, n_jobs=10, random_state=94)

model_a.fit(X_train_a, y_train)
pred_a = model_a.predict(X_test_a)

report_a = classification_report(y_test, pred_a, digits=3, output_dict=True)
print(classification_report(y_test, pred_a, digits=3))

precision_a = round(report_a["macro avg"]["precision"], 3)
recall_a = round(report_a["macro avg"]["recall"], 3)
f1_a = round(report_a["macro avg"]["f1-score"], 3)

precision_a, recall_a, f1_a
```

```python
df_b = df.copy()

df_b.iloc[:, 4] = df_b.iloc[:, 4].fillna("").astype(str)
df_b.iloc[:, 5] = df_b.iloc[:, 5].fillna("").astype(str)
df_b.iloc[:, 1] = pd.to_datetime(df_b.iloc[:, 1], errors="coerce")

text_b = (df_b.iloc[:, 4] + " " + df_b.iloc[:, 5]).str.strip()
weekday_b = df_b.iloc[:, 1].dt.weekday

x_b = pd.DataFrame({"weekday": weekday_b, "text": text_b})

x_train_b, x_test_b, _, _ = train_test_split(x_b, y, test_size=0.25, random_state=94)
```

```python
df_b = df.copy(deep=True)

for i in range(len(df_b)):
    if df_b.iat[i, 4] != "":
        df_b.iat[i, 5] = str(df_b.iat[i, 4]) + " " + str(df_b.iat[i, 5])

x_train_b, x_test_b, _, _ = train_test_split(
    df_b.iloc[:, [1, 5]], df_b.iloc[:, 0], test_size=0.25, random_state=94
)

vectorizer_b = TfidfVectorizer()

X_train_b = sp.sparse.hstack(
    (
        vectorizer_b.fit_transform(x_train_b.iloc[:, 1]),
        x_train_b.iloc[:, 0].values.reshape(len(x_train_b), 1),
    )
)

X_test_b = sp.sparse.hstack(
    (
        vectorizer_b.transform(x_test_b.iloc[:, 1]),
        x_test_b.iloc[:, 0].values.reshape(len(x_test_b), 1),
    )
)

model_b = RandomForestClassifier(n_estimators=8, n_jobs=10, random_state=94)

model_b.fit(X_train_b, y_train)
pred_b = model_b.predict(X_test_b)

report_b = classification_report(y_test, pred_b, digits=3, output_dict=True)
print(classification_report(y_test, pred_b, digits=3))

precision_b = round(report_b["macro avg"]["precision"], 3)
recall_b = round(report_b["macro avg"]["recall"], 3)
f1_b = round(report_b["macro avg"]["f1-score"], 3)

precision_b, recall_b, f1_b
```
