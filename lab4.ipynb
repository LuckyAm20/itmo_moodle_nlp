```python
import pandas as pd
import nltk
from nltk.corpus import wordnet as wn
from itertools import product
from scipy.stats import spearmanr
```

```python
nltk.download("wordnet")
```

```python
df = pd.read_csv("Task_4_sample_3.csv")
df.head()
```

```python
path_list = []
lch_list = []
wup_list = []
true_list = []

for _, row in df.iterrows():
    w1 = row["word_1"]
    w2 = row["word_2"]
    score = row["Score"]

    all_w1 = wn.synsets(w1, pos="n")
    all_w2 = wn.synsets(w2, pos="n")

    path_scores = []
    lch_scores = []
    wup_scores = []

    for s1, s2 in product(all_w1, all_w2):
        path_val = s1.path_similarity(s2)
        lch_val = None
        wup_val = s1.wup_similarity(s2)

        if path_val is not None:
            path_scores.append(path_val)

        try:
            lch_val = s1.lch_similarity(s2)
            if lch_val is not None:
                lch_scores.append(lch_val)
        except:
            pass

        if wup_val is not None:
            wup_scores.append(wup_val)

    path_list.append(max(path_scores) if path_scores else 0)
    lch_list.append(max(lch_scores) if lch_scores else 0)
    wup_list.append(max(wup_scores) if wup_scores else 0)
    true_list.append(score)
```

```python
path_corr, _ = spearmanr(path_list, true_list)
lch_corr, _ = spearmanr(lch_list, true_list)
wup_corr, _ = spearmanr(wup_list, true_list)

round(path_corr, 3), round(lch_corr, 3), round(wup_corr, 3)
```
